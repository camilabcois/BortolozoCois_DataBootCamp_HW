### Pyber – Report:

1)	My first observation is that it is more expensive to use Pyber in rural areas. Rural cities have the highest average fare. Also, this city type has the least number of drivers and rides.

2)	Second, even though rural areas have the highest average fare, the biggest percentage of total fare comes from urban cities.

3)	Finally, more than 80% of the Pyber drivers are from urban city type. 


```python
%matplotlib inline

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load 
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
combined_city_ride = pd.merge(ride_df, city_df, on= "city", how= "left")

# Display the data table for preview
combined_city_ride.head()
```

```python
# Create a Data Frame groupby city and type
grouped_city = combined_city_ride.groupby(["city", "type"])
grouped_city.head()
```

```python
# Calculate the average fare per city
avg_fare_city = grouped_city["fare"].mean()

# Calculate the total number rides per city
total_rides_city = grouped_city["ride_id"].count()

# Calculate the total number drivers per city
total_drivers_city = grouped_city["driver_count"].first()

# Colect each city type
city_type = combined_city_ride.type.unique()
```

## Bubble Plot of Ride Sharing Data

```python
# Create a summary table for ride sharing data
summary_city_df1 = pd.DataFrame({"total rides per city": total_rides_city,
                              "avg fare per city": avg_fare_city,
                              "total drivers per city": total_drivers_city})
# Reset the index
summary_city_df2 = summary_city_df1.reset_index()

# Show summary table
summary_city_df2.head()
```

```python
# Create a DataFrame for each city type
urban = summary_city_df2.loc[(summary_city_df2["type"] == "Urban")]
suburban = summary_city_df2.loc[(summary_city_df2["type"] == "Suburban")]
rural = summary_city_df2.loc[(summary_city_df2["type"] == "Rural")]

# Set the index as type 
urban_city = urban.set_index("type")
suburban_city = suburban.set_index("type")
rural_city = rural.set_index("type")

# Obtain the x and y coordinates and number of drivers for urban city type 
x_urban = urban_city.loc[:,"total rides per city"]
y_urban = urban_city.loc[:, "avg fare per city"]
drivers_urban = urban_city.loc[:,"total drivers per city"]

# Obtain the x and y coordinates and number of drivers for suburban city type 
x_suburban = suburban_city.loc[:,"total rides per city"]
y_suburban = suburban_city.loc[:, "avg fare per city"]
drivers_suburban = suburban.loc[:,"total drivers per city"]

# Obtain the x and y coordinates and number of drivers for rural city type 
x_rural = rural_city.loc[:,"total rides per city"]
y_rural = rural_city.loc[:,"avg fare per city"]
drivers_rural = rural.loc[:,"total drivers per city"]

# Create scatter plots for each city type
urban_scatter = plt.scatter(x_urban, y_urban, s=drivers_urban*10, marker="o", facecolors="coral", edgecolors="black", alpha = 0.5, label="urban")

suburban_scatter = plt.scatter(x_suburban, y_suburban, s=drivers_suburban*10, marker="o", facecolors="lightblue", edgecolors="black", alpha = 0.5, label="suburban")

rural_scatter = plt.scatter(x_rural, y_rural, marker="o", s=drivers_rural*10, facecolor="gold", edgecolors="black", alpha = 0.5, label="rural" )


# Add grid
plt.grid()

# Create a legend
plt.legend(loc='upper right', title = "City Types")

# Create a text label regarding circle size
plt.text(42, 35, "Note:\nCircle size correlates with driver count per city.")

# Add labels for axes and title
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")

# Save Figure
plt.savefig("../Images/bubble.png")


```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
fares_city_pie = combined_city_ride.groupby(["type"])["fare"].sum()

# Set labels
labels = ["Rural", "Suburban", "Urban"]

# Set title
plt.title("% of Total Fares by City Type")

# Set colors of each section of the pie chart
colors = ["gold", "lightblue", "lightcoral"]

# Tells matplotlib to seperate "Urban" section from the others
explode = (0 ,0 ,0.2 )

# Creates the pie chart based upon the values above
plt.pie(fares_city_pie, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)

# Save Figure
plt.savefig("../Images/pie1.png")
```

```python
# Show Figure
plt.show()
```
